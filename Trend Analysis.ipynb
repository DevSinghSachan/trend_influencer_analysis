```python
% reload_ext autoreload
% autoreload 2
% matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt, numpy as np, re, datetime
import utils, sys
sys.path.append("/Users/jonathanraiman/Desktop/Coding/language_modeling")
from word2vec_extended import Word2VecExtended as LanguageModel
lm = LanguageModel.load("/Users/jonathanraiman/Desktop/Coding/language_modeling/saves/kid_model_30_oov_senti")
MINCOUNT = 4
docs = utils.read_documents("data/cxs.11282014.dat.gz")
```

## Unigrams

Many of the words in our corpus are unspecific to the subject at hand. An easy way of finding specific and interesting unigrams is to look at the term frequency inverse document frequency, or **TF-IDF**. This metric assumes the probability of occurence of words is independent of the document. If the ratio is much higher than 1, than it appears that the document $d$ significantly boosted the likelihood of occurence of a word, and thus the document $d$ and this word $w_i$ may not be independent:

$s_{\mathrm{TF-IDF}} = \frac{\overbrace{P(w_i | d)}^{\text{document frequency}}}{\underbrace{P(w_i)}_{\text{term frequency}}}$


```python
counts = utils.get_word_counts(docs)
total_words = sum(val for val in counts.values() if val > MINCOUNT)
total_words_in_lm = sum(word.count for word in lm.vocab.values())
word_tf_idf = {}
deletes = []
for word in counts:
    try:
        tf = (lm.vocab[word].count / total_words_in_lm)
        df = counts[word] / total_words
        word_tf_idf[word] = df / tf
    except KeyError:
        if len(word) > 2 and counts[word] > MINCOUNT:
            df = counts[word] / total_words
            word_tf_idf[word] = df / 3.0
        else:
            deletes.append(word)
for delete in deletes:
    del counts[delete]
        
import operator
sorted_tf_idf = list(word_tf_idf.items())
sorted_tf_idf.sort(key=operator.itemgetter(1), reverse=True)
```

## Bi-Grams

Now we look at bigram occurence and probability. We can weed out bad bi-grams by supposing a Naive Bayes model for words:

$s = \frac{P( w_{i,j})}{P(w_i) \cdot P(w_j)}$

if the ratio $s$ is much higher than 1, then the probability of occurence of the pair of word $i$ and word $j$ is not independent, and thus we can consider $w_{i,j}$ to be a word of interest.


```python
bicounts = utils.get_bigram_counts(docs)
total_bicount = sum(val for val in bicounts.values() if val > MINCOUNT)
probed_bicounts = {}
for bicount in bicounts:
    if bicounts[bicount] > MINCOUNT and counts[bicount[0]] > 0 and counts[bicount[1]] > 0:
        f = bicounts[bicount] / total_bicount
        ind_f = (counts[bicount[0]] * counts[bicount[1]]) / (total_words * total_words)
        probed_bicounts[bicount] = f / ind_f
    
sorted_prob_bicount = list(probed_bicounts.items())
sorted_prob_bicount.sort(key=operator.itemgetter(1), reverse=True)
sorted_prob_bicount_np = np.array([[k, v] for k,v in sorted_prob_bicount])
```

As we can see in the plot below, the bigrams follow a power law for common bi-grams *Matthew McConaughey*, *Kip Thorne*, and then switch into a different dynamic, when the score $s$ dips below 100, as visible for phrases: *from the*, etc.:

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
def find_pos(el, arr):
    for k, el2 in enumerate(arr):
        if el == el2: return k
    return -1
ax.plot(np.arange(0, len(sorted_prob_bicount)), sorted_prob_bicount_np[:,1])
ax.set_yscale('log')
for word in [("Kip", "Thorne"), ("Matthew", "McConaughey"), ("from","the"), ("you", "re")]:
    ax.annotate(
        " ".join(word),
        xy= (find_pos(word, sorted_prob_bicount_np[:,0]), probed_bicounts[word]),
        xytext=(find_pos(word, sorted_prob_bicount_np[:,0]) + 500, probed_bicounts[word] * 2),xycoords='data', textcoords='data',
        arrowprops=dict(facecolor='#333333', width=0.7, shrink = 0.05, headwidth=8, linewidth=0));
plt.setp(ax.xaxis, visible=False);
ax.set_ylabel(r"$\frac{P( w_{i,j})}{P(w_i) \cdot P(w_j)}$");
ax.set_title("Log probability of bigrams");
```

```python
bigram_cutoff = 100
useful_bigrams = [[pair, bicounts[pair]] for pair in sorted_prob_bicount_np[np.nonzero(sorted_prob_bicount_np[:,1].astype(np.int32) > bigram_cutoff)[0],0]]
useful_bigrams.sort(key=operator.itemgetter(1), reverse=True)
useful_bigrams = np.array(useful_bigrams)

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
def find_pos(el, arr):
    for k, el2 in enumerate(arr):
        if el == el2: return k
    return -1
ax.plot(np.arange(0, len(useful_bigrams)), useful_bigrams[:,1])
ax.set_yscale('log')
for word in [("Kip", "Thorne"), ("Matthew", "McConaughey")]:
    ax.annotate(
        " ".join(word),
        xy= (find_pos(word, useful_bigrams[:,0]), bicounts[word]),
        xytext=(find_pos(word, useful_bigrams[:,0]) + 100, bicounts[word] * 2),xycoords='data', textcoords='data',
        arrowprops=dict(facecolor='#333333', width=0.7, shrink = 0.05, headwidth=8, linewidth=0));
for word_index in [10, 200, 300]:
    word = useful_bigrams[word_index,0]
    ax.annotate(
        " ".join(word),
        xy= (find_pos(word, useful_bigrams[:,0]), bicounts[word]),
        xytext=(find_pos(word, useful_bigrams[:,0]) + 100, bicounts[word] * 2),xycoords='data', textcoords='data',
        arrowprops=dict(facecolor='#333333', width=0.7, shrink = 0.05, headwidth=8, linewidth=0));
plt.setp(ax.xaxis, visible=False);
ax.set_ylabel("count");
ax.set_title("Log count for high scoring bigrams");
```

## Trends for unigrams and bigrams


Here we're looking at the days that cause peaks of occurence for a single unigram:

```python
word_trends = {}

test_word ="technology"

for doc in docs:
    date = datetime.date(doc.date.year, doc.date.month, doc.date.day)
    c = doc.content.count(test_word)
    if c > 0:
        if date in word_trends:
            word_trends[date] += c
        else:
            word_trends[date] = c

fig, ax = plt.subplots()

dates_counts = []
for key, val in word_trends.items():
    dates_counts.append([key, val])
dates_counts.sort(key=operator.itemgetter(0))
dates_counts = np.array(dates_counts)

go_back = 30 # days

ax.plot_date(dates_counts[:,0], dates_counts[:,1], '-');
ax.set_xlim([dates_counts[-1,0] - datetime.timedelta(go_back), dates_counts[-1,0]])
fig.autofmt_xdate()
ax.set_ylabel("Occurence of \"%s\"" % (test_word));
ax.set_title("Evolution of unigram occurence");
```

### Generalizing the technique

Now we want to study multiple trends in parallel

```python
def show_bigram_trends(bigrams, multiply_by_audience= False, window = 30, log_audience = False):

    word_trends = {bigram: {} for bigram in bigrams}

    for doc in docs:
        date = datetime.date(doc.date.year, doc.date.month, doc.date.day)
        split_doc = " ".join([w for w in re.split(utils.splitter, doc.content) if len(w) > 1 and w != " "])
        if multiply_by_audience:
            for bigram in bigrams:
                c = split_doc.count(bigram[0]+" "+bigram[1]) * (doc.count if not log_audience else np.log1p(log_audience))
                if c > 0:
                    if date in word_trends[bigram]:
                        word_trends[bigram][date] += c
                    else:
                        word_trends[bigram][date] = c
        else:
            for bigram in bigrams:
                c = split_doc.count(bigram[0]+" "+bigram[1])
                if c > 0:
                    if date in word_trends[bigram]:
                        word_trends[bigram][date] += c
                    else:
                        word_trends[bigram][date] = c


    fig, ax = plt.subplots()
    latest = None
    for bigram in bigrams:
        if len(word_trends[bigram]) >0:
            dates_counts = []
            for key, val in word_trends[bigram].items():
                dates_counts.append([key, val])
            dates_counts.sort(key=operator.itemgetter(0))
            dates_counts = np.array(dates_counts)

            ax.plot_date(dates_counts[:,0], dates_counts[:,1], '-');
            if latest is None or dates_counts[-1,0] > latest:
                latest = dates_counts[-1,0]
    ax.set_xlim([latest - datetime.timedelta(window), latest])
    fig.autofmt_xdate()
    ax.set_ylabel("Occurence" if not multiply_by_audience else "Occurence * audience");
    ax.set_title("Evolution of occurence");
```
